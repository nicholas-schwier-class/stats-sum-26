```python
#Rolling two dice
import numpy as np

die1 = np.random.randint(1,7,10000)
die2 = np.random.randint(1,7,10000)

sums = die1 + die2

count = np.sum(sums == 7)

print(count / len(sums))
```

```python
#1000 Coin Tosses
import numpy as np

tosses = np.random.choice(
    ["Heads","Tails"],
    size=1000
)

heads = np.sum(tosses == "Heads")

print(heads)
```

```python
#ESTIMATING PROBABLITY
prob_heads = heads / len(tosses)

print(prob_heads)
```

Monte Carlo Simulation

Monte Carlo methods:

Use repeated random sampling to estimate results.

Applications:

Finance
Engineering
AI
Physics

```python
import numpy as np

trials = 100000

success = 0

for i in range(trials):

    rolls = np.random.randint(1,7,3)

    if 6 in rolls:
        success += 1

print(success/trials)
```
