What is Pandas?

Pandas is a Python library used for:

- Data analysis
- Data cleaning

Pandas is built on top of NumPy.

Common uses:

- Working with spreadsheets
- Analyzing survey results
- Processing business data
- Exploring scientific datasets

```python
#CREATING FIRST DATAFRAME

import pandas as pd

students = {
    "Name": ["Alice", "Bob", "Carol"],
    "Age": [20, 21, 19],
    "Major": ["Biology", "Math", "Psychology"]
}

df = pd.DataFrame(students)

print(df)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#Selecting a COLUMN
df["Name"]
```

```python
df[["Name", "Age"]]
```

```python
#Accessing Specific Rows with .loc
df.loc[1]
```

```python
#SELECTING ROW AND COLUMN
df.loc[1, "Major"]
```

```python
#Adding New Column
df["GPA"] = [3.5, 3.8, 3.7]

print(df)
```

```python
#DIFFERENCE BETWEEN DISPLAY AND PRINT
print(df)
display(df)
```

```python
grades = pd.DataFrame({
    "Student": ["A","B","C","D","E"],
    "Score": [78,85,92,88,95]
})

grades
```

```python
#Descriptive Statistics
grades.describe()
```

```python
#Filtering Data

#Find students scoring above 85:

grades[grades["Score"] > 85]
```

```python
#Multiple Condiditions
grades[
    (grades["Score"] > 80) &
    (grades["Score"] < 95)
]
```

```python
#sorting
grades.sort_values("Score")
```
