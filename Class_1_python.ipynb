```python
scores = [72, 75, 78, 80, 85, 85, 88, 90, 95]

#Questions:

#What is a typical score?
#How spread out are the scores?
#Are the scores consistent?
```

```python
import statistics

scores = [72, 75, 78, 80, 85, 85, 88, 90, 95]

statistics.mean(scores)
```

```python
statistics.median(scores)
```

```python
statistics.mode(scores)
```

image.png

```python
#Range
max(scores) - min(scores)
```

```python
#variance
statistics.variance(scores)
```

```python
#SD
statistics.stdev(scores)

```

```python
import matplotlib.pyplot as plt

plt.hist(scores)
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores")

plt.show()
```
